## 3. Data Cleaning

Removing asterisk from id

```python
cleaned_data['id'] = dirty_df['id'].replace({'\*':""},regex = True)
cleaned_data.head(10)
```

Converting case

```python
cleaned_data['neighbourhood']= dirty_df['neighbourhood'].str.title()
cleaned_data.head(10)
```

Mispelling mistakes fixed

```python
cleaned_data['neighbourhood_group'] = dirty_df['neighbourhood_group'].replace({'Brooklynn':"Brooklyn"},regex = True)
cleaned_data.head(10)
```

```python
cleaned_data['room_type'] = dirty_df['room_type'].replace({'Privaate room':"Private room"},regex = True)
cleaned_data.head(10)
```
