### ● What’s the average room price in each borough? 

```python
dd.columns
```

```python
pivot_dd = pd.pivot_table(dd, index='room_type', values='price', aggfunc=('min', 'median', 'max'))      
pivot_dd
```

```python
pivot_dd = dd.pivot_table('price', (['room_type' , 'neighbourhood_group']), aggfunc = ('mean', 'max', 'min'))      
pivot_dd
```
